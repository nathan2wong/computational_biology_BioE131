```python
from Bio import SeqIO
record = SeqIO.read("candidatus_solibacter.fa", "fasta")
```

```python
import csv
seqDic = {}
counter = 0;

with open('temp22M.fa') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    for row in csv_reader:
        cur = row[0]
        arr = cur.split("NODE_")
        if len(arr) > 1:
            node = arr[1].split("_")[0]
            seq = arr[1].split("\t")[2]
            seqDic[node] = seq
```

```python
keys = [];
for i in seqDic:
    keys.append(int(i.split()[0]))
```

```python
def mergeSort(arr): 
    if len(arr) >1: 
        mid = len(arr)//2 #Finding the mid of the array 
        L = arr[:mid] # Dividing the array elements  
        R = arr[mid:] # into 2 halves 
  
        mergeSort(L) # Sorting the first half 
        mergeSort(R) # Sorting the second half 
  
        i = j = k = 0
          
        # Copy data to temp arrays L[] and R[] 
        while i < len(L) and j < len(R): 
            if L[i] < R[j]: 
                arr[k] = L[i] 
                i+=1
            else: 
                arr[k] = R[j] 
                j+=1
            k+=1
          
        # Checking if any element was left 
        while i < len(L): 
            arr[k] = L[i] 
            i+=1
            k+=1
          
        while j < len(R): 
            arr[k] = R[j] 
            j+=1
            k+=1
  
# Code to print the list 
def printList(arr): 
    for i in range(len(arr)):         
        print(arr[i],end=" ") 
    print() 
```

```python
print ("Given array is", end="\n")  
mergeSort(keys) 
print("Sorted array is: ", end="\n") 
printList(keys)
```

```python
fa = []
for k in keys:
    fa.append(seqDic[str(k)])

print(fa)
```

```python
with open('combined.fa', 'w') as f:
    for item in fa:
        f.write("%s\n" % item)
```
