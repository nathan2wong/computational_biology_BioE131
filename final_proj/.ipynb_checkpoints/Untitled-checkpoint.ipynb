```python
import urllib.request, json 
NOG_processes = []
NOG_nums = []
Subsystems_processes = []
Subsystems_nums = []
COG_processes = []
COG_nums = []
KO_processes = []
KO_nums = []
newlist = list()

with urllib.request.urlopen("https://api-ui.mg-rast.org/metagenome/26355e360b6d676d343832313537352e33?verbosity=stats&detail=ontology&auth=qguPGuFpjwq3e5B4PSZPsVTa3") as url:
    data = json.loads(url.read().decode())
    for i in data.keys():
        newlist.append(i)
    for i in range(len(data['NOG'])):
        NOG_processes.append(data['NOG'][i][0])
        NOG_nums.append(data['NOG'][i][1])
    for i in range(len(data['Subsystems'])):
        Subsystems_processes.append(data['Subsystems'][i][0])
        Subsystems_nums.append(data['Subsystems'][i][1])
    for i in range(len(data['COG'])):
        COG_processes.append(data['COG'][i][0])
        COG_nums.append(data['COG'][i][1])
    for i in range(len(data['KO'])):
        KO_processes.append(data['KO'][i][0])
        KO_nums.append(data['KO'][i][1])

print(newlist)
```

```python
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt

objects = tuple(NOG_processes)
y_pos = np.arange(len(objects))
performance = NOG_nums
 
plt.barh(y_pos, performance, align='center', alpha=0.5, color=['black', 'red', 'green', 'blue'])
plt.yticks(y_pos, objects)
plt.xlabel('Times Appeared')
plt.title('NOG')

plt.show()

objects = tuple(Subsystems_processes)
y_pos = np.arange(len(objects))
performance = Subsystems_nums
 
plt.barh(y_pos, performance, align='center', alpha=0.5)
plt.yticks(y_pos, objects)
plt.xlabel('Times Appeared')
plt.title('Subsystems')
 
plt.show()

objects = tuple(COG_processes)
y_pos = np.arange(len(objects))
performance = COG_nums
 
plt.barh(y_pos, performance, align='center', alpha=0.5)
plt.yticks(y_pos, objects)
plt.xlabel('Times Appeared')
plt.title('COG')
 
plt.show()

objects = tuple(KO_processes)
y_pos = np.arange(len(objects))
performance = KO_nums

plt.barh(y_pos, performance, align='center', alpha=0.5)
plt.yticks(y_pos, objects)
plt.xlabel('Times Appeared')
plt.title('KO')
 
plt.show()
```

```python
import pandas
import matplotlib.pyplot as plt
import numpy as np

df = pandas.DataFrame(dict(graph=['Item one', 'Item two', 'Item three'],
                           n=[3, 5, 2], m=[6, 1, 3])) 

ind = np.arange(len(df))
width = 0.4

fig, ax = plt.subplots()
ax.barh(ind, df.n, width, color='red', label='Layer 1')
ax.barh(ind + width, df.m, width, color='green', label='Layer 2')

ax.set(yticks=ind + width, yticklabels=df.graph, ylim=[2*width - 1, len(df)])
ax.legend()

plt.show()
```
