```python
allGenus = []
allOccurs = []
```

```python
import json
import pandas as pd

genus = []
url = []
occur = []
genusDic = {}
Jfile = "layer1.json"

with open(Jfile) as f:
    data = json.load(f)

for tab in data:
    genus.append(tab['label'])
    occur.append(tab['value'])
    #url.append(tab['click'].split("open(")[1].split(");")[0])
    genusDic[tab['label']] = tab['value']
    #print(tab)
print(genusDic)
```

```python
#Data post process
occur = [g for g in occur if g > 500]
genus = []
for i in genusDic:
    if(genusDic[i] in occur):
        genus.append(i)
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
plt.title("Position in the genome vs. Depth of coverage")
plt.ylabel('Depth of Coverage')
plt.xlabel('Postion in Genome')
plt.scatter(x = genus, y = occur, s = 1, color = 'red')
plt.show()
```

```python
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = occur
explode = (0, 0.1, 0, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')
allGenus.append(genus)
allOccurs.append(occur)
mpl.rcParams['font.size'] = 50.0
fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=genus, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.figure(figsize=(2000,1001))
fig1.set_size_inches(50,50)
plt.show()
fig1.savefig(Jfile+".png")
```
