<h1>1. Read Text File</h1>

```python
sentences = []
with open('romeoandjuliet.txt') as RJText:
    data = RJText.readlines()
    for x in data:
        sentences += x.split()
```

<h1>2. Create Dictionary with unique words as keys. Values of each key are how many times (frequency) each word appears in the data file. </h1>

```python
#Construct dic
import string
exclude = set(string.punctuation)
words = {}
for word in sentences:
    word = ''.join(ch for ch in word if ch not in exclude).lower()
    if word in words:
        words[word] += 1
    else:
        words[word] = 1
```

<h1>3.1 Sort the dictionary in decreasing value (Greatest word count first). Graph of all words and their frequency is plotted. </h1>

```python
#Word decreasing sort
sortedwords = sorted(words, key=words.get, reverse=True)

from matplotlib import pyplot as plt
plt.bar(range(len(words)), list(words.values()), align='center')
plt.xticks(range(len(words)), list(words.keys()))
plt.show()
```

<h1>3.2 Plot the top 10 most frequent words.</h1>

```python
#Top Ten
topten = sortedwords[:10]
toptendic={}
for word in topten:
    toptendic[word] = words[word]
print(toptendic)

plt.title("Top 10 most frequent words")
plt.bar(range(len(toptendic)), list(toptendic.values()), align='center')
plt.xticks(range(len(toptendic)), list(toptendic.keys()))
plt.show()
```

<p>Most common word used by Shakespeare in Romeo and Juliet appear to be "and" at 713 times. The next most common word is "the" at 680 times. This means that most of words are used not descriptive words or verbs, but rather connecting and supporting words</p>

<h1>3.3 Least Frequent words plotted (appears only once).</h1>

```python
#Least Frequent
leastfrequent = [word for word in words if words[word] == 1]

leastfrequentdic={}
for word in leastfrequent:
    leastfrequentdic[word] = words[word]

plt.bar(range(len(leastfrequentdic)), list(leastfrequentdic.values()), align='center')
plt.xticks(range(len(leastfrequentdic)), list(leastfrequentdic.keys()), size='small')
plt.show()

print(len(leastfrequent))
print(len(words))
print(leastfrequent)
```

<p>Least common words used by Shakespeare in Romeo and Juliet are words he only uses once in the entire novel. An example of these words is "encamp: and "wentst". There are 2087 words he only uses once. One conclusion to draw from this is that Shakespeare uses a very wide vocabulary in his novels, as so many of the words he chooses to use only appear once. In fact, of the 3764 words used in Romeo Juliet, <strong>around 55% of them are unique.</strong></p>

<h1>3.4/4 Sorted Dictionary (from step 3.1). Plot of word count vs word. Log of word count vs word.

```python
#Most frequent LOG
import math
NUM_WORDS = 20

frequent = sortedwords[:NUM_WORDS]
frequentval = []

i = 0
while i < NUM_WORDS:
    frequentval.append(words[frequent[i]])
    i+= 1
    
plt.title(s="Most frequent words")
plt.bar(frequent, frequentval, align='center')
plt.xticks(frequent, frequent, size="small")
plt.show()
    
    
plt.title(s="Log of most frequent words")
plt.bar(frequent, [math.log(x) for x in frequentval], align='center')
plt.xticks(frequent, frequent, size="small")
plt.show()
```

<h1>4. Plot of word rank vs word Frequency. Logarthmic scale implemented as well.

```python
import numpy as np

NUM_WORDS = 20

frequent = sortedwords[:NUM_WORDS]
frequentval = []
xindex = [i for i in range(0, NUM_WORDS)]

i = 0
while i < NUM_WORDS:
    frequentval.append(words[frequent[i]])
    i+= 1

#Regular no log top 20
plt.title(s="Word frequency vs word rank top 20")
plt.plot(frequentval)
plt.xticks(np.arange(1,NUM_WORDS))
plt.show()

#With log top 20
plt.title(s="Logarthmic Word frequency vs word rank top 20")
plt.plot(xindex,[math.log(x) for x in frequentval])
plt.xticks(xindex, [round(math.log(x)) for x in np.arange(1,NUM_WORDS)])
plt.show()
```

```python
#Top 50
NUM_WORDS = 50

frequent = sortedwords[:NUM_WORDS]
frequentval = []
xindex = [i for i in range(0, NUM_WORDS)]

i = 0
while i < NUM_WORDS:
    frequentval.append(words[frequent[i]])
    i+= 1

#Regular no log top 50
plt.title(s="Word frequency vs word rank top 50")
plt.plot(frequentval)
plt.xticks(np.arange(1,NUM_WORDS))
plt.show()

#With log top 50
plt.title(s="Logarthmic Word frequency vs word rank top 50")
plt.plot(xindex,[math.log(x) for x in frequentval])
plt.xticks(xindex, [round(math.log(x)) for x in np.arange(1,NUM_WORDS)])
plt.show()
```

```python
#All words
NUM_WORDS = len(sortedwords)

frequent = sortedwords[:NUM_WORDS]
frequentval = []
xindex = [i for i in range(0, NUM_WORDS)]

i = 0
while i < NUM_WORDS:
    frequentval.append(words[frequent[i]])
    i+= 1

#Regular no log all
plt.title(s="Word frequency vs word rank top 20")
plt.plot(frequentval)
plt.xticks(np.arange(1,NUM_WORDS))
plt.show()

#With log all
plt.title(s="Logarthmic Word frequency vs word rank top 20")
plt.plot(xindex,[math.log(x) for x in frequentval])
plt.xticks(xindex, [round(math.log(x)) for x in np.arange(1,NUM_WORDS)])
plt.show()
```

<h4>By looking at the word frequency used by Shakespeare and its logarithmic scale, it is clear that there are a lot of words Shakespeare uses many times, and a lot of words Shakespeare uses very few times, but hardly any words that Shakespeare uses an average amount of time.</h4>
